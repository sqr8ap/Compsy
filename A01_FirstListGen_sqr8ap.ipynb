```python
import random
import pickle

coherence_levels = [0.0, 0.1, 0.2, 0.3]
inter_times = [200, 300, 500, 800]
num_reps = 2
num_blocks = 2
```

```python
def gen_conds():
    """
    This function must be called to generate the conditions for this experiment.
    It takes no input but relies on global variables coherence_levels and inter_times. 
    It returns the conditions as a list of dictionaries.
    """
    global inter_times
    global coherence_levels
    conds = []
    for time in inter_times:
        trial = {"left_coh": 0, "right_coh": 0, "inter_time": time}
        for level in coherence_levels:
            trial = {"left_coh": level, "right_coh": 0, "inter_time": time}
            for level2 in coherence_levels:
                trial['right_coh'] = level2
                conds.append(trial)
                trial = {"left_coh": level, "right_coh": level2, "inter_time": time}
    return conds

gen_conds()

def gen_trials(num_reps):
    """
    This function generates the full set of trials.
    It takes num_reps as input and generates randomly ordered repetitions of the conditions.
    It returns the list of trials with each unique condition repeated num_reps number of times. 
    """
    trials = conds * num_reps
    random.shuffle(trials)
    return trials

blocks = [gen_trials(num_reps) for i in range(num_blocks)] # because gen_trials is called for each block, the blocks will each have shuffled orders
```

```python
pickle.dump(blocks, open('blocks_pickle.pickle', 'wb'))
```
